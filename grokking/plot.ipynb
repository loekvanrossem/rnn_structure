```python
import sys
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

source = "../source"
sys.path.append(source)

import publication
```

```python
df = pd.read_csv("run_data_4.csv")

# run_name = "atomic-glade-337"
# eval_every = 500

# run_name = "noble-moon-342"
# eval_every = 250

# run_name = "misty-eon-360"
# eval_every = 250

run_name = "hopeful-waterfall-375"
eval_every = 250

steps = df["Step"]
epochs = steps * eval_every
train_loss = df[f"{run_name} - train.loss"]
val_loss = df[f"{run_name} - val.loss"]
train_acc = df[f"{run_name} - train.accuracy"]
val_acc = df[f"{run_name} - val.accuracy"]
n_states_hidden = df[f"{run_name} - n_states_hidden"]
n_states_attn = df[f"{run_name} - n_states_attn"]
n_states_keys = df[f"{run_name} - n_states_keys"]
n_states_vals = df[f"{run_name} - n_states_vals"]
# epochs = np.log(epochs)
```

```python
plt.figure(figsize=(6, 3))
publication.set_color_mixed(0)
c = 10

plt.plot(epochs, train_loss / c, label="train loss")
plt.plot(epochs, val_loss / c, label="val loss")
plt.plot(epochs, n_states_attn, label="number of states (attention)", zorder=2)
publication.set_color_mixed(1)
plt.gca().set_prop_cycle(None)
publication.skip_colors(2)
plt.plot(epochs, n_states_hidden, label="number of states (hidden)", zorder=1)

plt.xlabel("Epochs")
plt.ylim(0, np.max(val_loss / c))
# plt.legend()
publication.plt_show(save_path="number_of_states_grokking")
```

```python
plt.figure(figsize=(6, 3))
publication.set_color_mixed(0)
c = 100

plt.plot(epochs, train_loss / c, label="train loss")
plt.plot(epochs, val_loss / c, label="val loss")
plt.plot(epochs, n_states_keys, label="number of states (keys)", zorder=2)
publication.set_color_mixed(1)
plt.gca().set_prop_cycle(None)
publication.skip_colors(2)
plt.plot(epochs, n_states_vals, label="number of states (values)", zorder=1)

plt.xlabel("Epochs")
plt.ylim(0, np.max(n_states_vals))
# plt.legend()
publication.plt_show(save_path="number_of_states_grokking_keys_vals")
```
