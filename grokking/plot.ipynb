```python
df = pd.read_csv("run_data.csv")
run_name = "atomic-glade-337"
eval_every = 500

steps = df["Step"]
epochs = steps * eval_every
train_loss = df[f"{run_name} - train.loss"]
val_loss = df[f"{run_name} - val.loss"]
train_acc = df[f"{run_name} - train.accuracy"]
val_acc = df[f"{run_name} - val.accuracy"]
n_states_hidden = df[f"{run_name} - n_states_hidden"]
n_states_attn = df[f"{run_name} - n_states_attn"]
```

```python
plt.figure(figsize=(4, 3))
publication.set_color_mixed(0)
plt.plot(epochs, train_loss / 10, label="train loss")
plt.plot(epochs, val_loss / 10, label="val loss")
plt.plot(epochs, n_states_attn, label="number of states (attention)", zorder=2)
publication.set_color_mixed(1)
plt.gca().set_prop_cycle(None)
publication.skip_colors(2)
plt.plot(epochs, n_states_hidden, label="number of states (hidden)", zorder=1)

plt.xlabel("Epochs")
plt.ylim(0, 1.1)
# plt.legend()
publication.plt_show(save_path="number_of_states_grokking")
```
